# MNIST Handwritten Digit Recognition
Alternative Models

```

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```

X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
X = X / 255.0
y = y.astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```

def evaluate(model, X_test, y_test, title):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{title} Accuracy = {acc:.4f}")
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    return acc
```

```

knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train, y_train)
acc_knn5 = evaluate(knn5, X_test, y_test, "KNN (k=5)")

knn9 = KNeighborsClassifier(n_neighbors=9, weights='distance')
knn9.fit(X_train, y_train)
acc_knn9 = evaluate(knn9, X_test, y_test, "KNN (k=9, distance)")
```

```

X_train_svm = X_train[:10000]
y_train_svm = y_train[:10000]

svm_lin = SVC(kernel='linear', C=1)
svm_lin.fit(X_train_svm, y_train_svm)
acc_svm_lin = evaluate(svm_lin, X_test, y_test, "SVM Linear")

svm_rbf = SVC(kernel='rbf', C=5)
svm_rbf.fit(X_train_svm, y_train_svm)
acc_svm_rbf = evaluate(svm_rbf, X_test, y_test, "SVM RBF")
```

```

dt10 = DecisionTreeClassifier(max_depth=10)
dt10.fit(X_train, y_train)
acc_dt10 = evaluate(dt10, X_test, y_test, "Decision Tree (depth=10)")

dt20 = DecisionTreeClassifier(max_depth=20)
dt20.fit(X_train, y_train)
acc_dt20 = evaluate(dt20, X_test, y_test, "Decision Tree (depth=20)")
```

```

import pandas as pd

results = pd.DataFrame({
    "Model": [
        "KNN k=5", "KNN k=9",
        "SVM Linear", "SVM RBF",
        "Decision Tree depth=10", "Decision Tree depth=20"
    ],
    "Accuracy": [
        acc_knn5, acc_knn9,
        acc_svm_lin, acc_svm_rbf,
        acc_dt10, acc_dt20
    ]
})

results
```
